```python
import numpy as np
from matplotlib import pyplot as plt
from sklearn.neighbors import KNeighborsRegressor
from sklearn.base import BaseEstimator, RegressorMixin
from sklearn.utils import check_array, check_X_y
from sklearn.metrics import pairwise_distances
from sklearn.model_selection import cross_val_score
```

## Kernel Regression

```python
class KernelRegressor(BaseEstimator, RegressorMixin):
    def __init__(self, K=None, h=None):
        self.K = K
        self.h = h
        
    def fit(self, X, y):
        _X, _y = check_X_y(X, y)
        self._X = _X
        self._y = _y

    
    def predict(self, X):
        X = check_array(X)
        dists = self.K(pairwise_distances(X, self._X)/self.h)
        num = np.sum(dists*self._y.reshape(1, -1), axis=1)
        denom = np.sum(dists, axis=1)
        return num/denom
    
def B(u):
    return 0.5*(np.abs(u) < 1)

def T(u):
    return (1 - np.abs(u))*(np.abs(u) < 1)

def G(u):
    return np.exp(-np.square(u)/2)/np.sqrt(2*np.pi)
```

```python
t = np.linspace(-2, 2, 100)
plt.plot(t, B(t))
plt.plot(t, T(t))
plt.plot(t, G(t))
```

```python
np.random.seed(668)

def y_true(x):
    return np.sin(x*(2*np.pi*x))

t = np.linspace(0, 1)

n = 50
X = np.random.uniform(size=n)
y = y_true(X) + np.random.normal(0, 0.2, size=n)

plt.plot(t, y_true(t))
plt.plot(X, y, 'o')
```

```python
scores = []
hs = np.arange(0.01, 0.2, 0.02)
for h in hs:
    M = KernelRegressor(K=G, h=h)
    score = np.mean(-cross_val_score(M, X.reshape(-1, 1),  y, cv=5, scoring="neg_mean_squared_error"))
    scores.append(score)

# plt.plot(hs, scores)

h_opt = hs[np.argmin(scores)]
print(f"h_opt = {h_opt}")

M = KernelRegressor(K=G, h=h_opt)
M.fit(X.reshape(-1, 1), y)
y_p_G = M.predict(t.reshape(-1, 1))

scores = []
hs = np.arange(0.05, 0.2, 0.01)
for h in hs:
    M = KernelRegressor(K=T, h=h)
    score = np.mean(-cross_val_score(M, X.reshape(-1, 1),  y, cv=5, scoring="neg_mean_squared_error"))
    scores.append(score)

# plt.plot(hs, scores)

h_opt = hs[np.argmin(scores)]
print(f"h_opt = {h_opt}")

M = KernelRegressor(K=T, h=h_opt)
M.fit(X.reshape(-1, 1), y)
y_p_T = M.predict(t.reshape(-1, 1))

scores = []
hs = np.arange(0.07, 0.2, 0.01)
for h in hs:
    M = KernelRegressor(K=B, h=h)
    score = np.mean(-cross_val_score(M, X.reshape(-1, 1),  y, cv=5, scoring="neg_mean_squared_error"))
    scores.append(score)

# plt.plot(hs, scores)

h_opt = hs[np.argmin(scores)]
print(f"h_opt = {h_opt}")

M = KernelRegressor(K=B, h=h_opt)
M.fit(X.reshape(-1, 1), y)
y_p_B = M.predict(t.reshape(-1, 1))

scores = []
ks = np.arange(1, 20)
for k in ks:
    M = KNeighborsRegressor(n_neighbors=k)
    score = np.mean(-cross_val_score(M, X.reshape(-1, 1),  y, cv=5, scoring="neg_mean_squared_error"))
    scores.append(score)

# plt.plot(hs, scores)

k_opt = ks[np.argmin(scores)]
print(f"k_opt = {k_opt}")

M = KNeighborsRegressor(n_neighbors=k_opt)
M.fit(X.reshape(-1, 1), y)
y_p_N = M.predict(t.reshape(-1, 1))

fig, [[a0, a1], [a2, a3]]= plt.subplots(2, 2, figsize=(20, 10))
a0.plot(t, y_true(t))
a0.plot(t, y_p_G, lw=5)

a1.plot(t, y_true(t))
a1.plot(t, y_p_T, lw=5)

a2.plot(t, y_true(t))
a2.plot(t, y_p_B, lw=5)

a3.plot(t, y_true(t))
a3.plot(t, y_p_N, lw=5)
```
